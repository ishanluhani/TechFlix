```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('Movies (1970-2023).csv')
movies.head()
```

```python
# title_x
# imdb_id
# genres
# summary
# actors

movies = movies[['movie title', 'imdb id', 'plot', 'genres', 'cast', 'language cinema']]
movies.head()
```

```python
movies.columns
```

```python
def convert_to_list(column, till=None):
    print(column)
    print('-------------------------------------------------------')
    categories = column.split('|')
    if till:
        categories = categories[:till]
    return categories
```

```python
movies.dropna(inplace=True)
```

```python
from ast import literal_eval

movies['cast'] = movies['cast'].apply(lambda x: literal_eval(x)[:3])
movies['genres'] = movies['genres'].apply(lambda x: literal_eval(x))
movies.head()
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies['plot'] = movies['plot'].apply(str.split)
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [''.join(i.split()).lower() for i in x])
# movies['plot'] = movies['plot'].apply(lambda x: [''.join(i.split()) for i in x])
# movies['genres'] = movies['genres'].apply(lambda x: [''.join(i.split()) for i in x])
movies.head(1)
```

```python
movies['tags'] = movies['plot'] + movies['genres'] + movies['cast'] + movies['language cinema'].apply(lambda x: [x])
movies['tags'] = movies['tags'].apply(lambda x: [i.lower() for i in x])
movies.iloc[0]['tags']
```

```python
movies.head(1)
```

```python
new_movies = movies.drop(['plot', 'genres', 'cast'], axis=1)
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: ' '.join(x))
new_movies['tags'][0]
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()

def stem_columns(column):
    words = column.split()
    result = []
    for word in words:
        result.append(ps.stem(word))
    return ' '.join(result)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem_columns)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words='english', max_features=1000)
```

```python
mappings = vectorizer.fit_transform(new_movies['tags']).toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simularity = cosine_similarity(mappings)
```

```python
sorted(list(enumerate(simularity[0])), reverse=True, key=lambda x: x[1])
```
